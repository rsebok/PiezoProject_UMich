```python
import pandas as pd
from matplotlib import pyplot as plt
import csv
```

```python
# Load data and add column names
headers = ['X (mm)', 'Y (mm)']
data = pd.read_csv('.\Results\PrecisionXYTestNC_20230119_143758.csv', skiprows=1, names=headers) #select your file
#destination = pd.read_csv('.\Results\pre_20230113_145250.csv', nrows=1, names=headers) #select your file
start = pd.read_csv('.\Results\PrecisionXYTestNC_20230119_143758.csv', names=headers, skiprows=lambda x: (x != 0) and not x % 2)
end = pd.read_csv('.\Results\PrecisionXYTestNC_20230119_143758.csv', names=headers, skiprows=lambda x: x % 2)
```

```python
# Plot
steps = list(range(1,len(data['X (mm)'])+1))
figure = plt.figure(figsize=(15, 15))
ax = figure.add_subplot(111)
plt.scatter(start['X (mm)'], start['Y (mm)'],color='blue')
plt.scatter(end['X (mm)'], end['Y (mm)'],color='red')
plt.scatter(end['X (mm)'][0], end['Y (mm)'][0],color='orange')
circle1 = plt.Circle( (5.415049495711354, 6.654107624668449 ), 2.3577665518872997 , fill = False )
circle2 = plt.Circle( (5.415049495711354, 6.654107624668449 ), 0.25 , fill = False, color = 'r', linestyle = '--' )
ax.add_artist( circle1 )
ax.add_artist( circle2 )
plt.text(data['X (mm)'][0],data['Y (mm)'][0],'  START',size = 10)
plt.text(end['X (mm)'][-1:],end['Y (mm)'][-1:],'  END',size = 10)
plt.text(end['X (mm)'][0],end['Y (mm)'][0],'  GOAL',size = 10)
plt.plot(data['X (mm)'],data['Y (mm)'])
plt.title('Piezo CSV plot')
plt.xlim(2,8)
plt.ylim(4,10)
plt.xlabel('X (mm)')
plt.ylabel('Y (mm)')
plt.show()
```

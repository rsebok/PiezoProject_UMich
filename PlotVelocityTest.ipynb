```python
#load data
pointlist = 'AutoVelocityTest_20230605_113337'
```

```python
headers = ['X (mm)', 'Y (mm)']
data = pd.read_csv('.\Results\%s.csv' %pointlist, names=headers) #select your file
start = data.iloc[0::2, :]
end = data.iloc[1::2, :]
```

```python
print(start)
```

```python
endx = end['X (mm)'].tolist()
endy = end['Y (mm)'].tolist()
startx = start['X (mm)'].tolist()
starty = start['Y (mm)'].tolist()

xdiffs = np.subtract(endx, startx)
xdiffs = list(xdiffs)
xdiffs = [ele for ele in xdiffs if abs(ele) < 2]

ydiffs = np.subtract(endy, starty)
ydiffs = list(ydiffs)
ydiffs = [ele for ele in ydiffs if abs(ele) < 2]
```
